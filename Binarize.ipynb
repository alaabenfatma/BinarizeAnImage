### An approach to binarize an image using basic operations

The idea is to read an image as is, convert it into grayscale, and then convert the values depending on a specific value. 
>The value will be equal to 100 by default

Load the libraries.

*cv2* is an abbrevation for OpenCV. It will be used in order to read the image

*numpy* is the state-of-the-art library when it comes to working with arrays and matrices.

*matplotlib* is for visualization.

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Read the images
image = cv2.imread("pika.png",1)
print("Input image")
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
def binarize(img):
    img[img<150] = 255
    img[img<255] = 0
    return img
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = binarize(image)
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```
